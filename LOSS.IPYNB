```python
def gradien_descent(x,y):
    w_old=b_old=0
    w_values=
    cost_values=
    learning_rate=0.01
    n=len(x)
    iterations=100
    for i in range(iterations):
        y_predicted=w_old*x+b_old

        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        w_derivatives=-(2/n)*sum(x*(y-y_predicted))
        b_derivatives=-(2/n)*sum(x*(y-y_predicted))
        w_new=w_old-learning_rate*w_derivatives
        b_new=b_old-learning_rate*w_derivatives
        w_old=w_new
        b_old=b_new
        w_values.append(w_new)
        cost_values.append(cost)
        print("w={},b={},cost={},iterations={}".format(w_new,b_new,cost,i))

        
```

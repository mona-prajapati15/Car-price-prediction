```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
x,y=make_moons(n_samples=300,noise=0.05)
```

```python
db=DBSCAN(eps=0.02,min_samples=10,metric='euclidean')
```

```python
db.fit(x)
```

```python
label=db.labels_
```

```python
label
```

```python
plt.scatter(x[:, 0], x[:, 1], c=label)
```

```python
silhouette_score(x, label)
```

```python
from sklearn.model_selection import GridSearchCV #CV=Cross validation
import numpy as np
param={
    'eps':[0,45,16,41,23,12],
    'min_samples':[6,10,13,12]
}
```

```python
db1=DBSCAN()
```

```python
GridSearchCV(db1,param_grid=param,cv=5,scoring=silhouette_score)
```
